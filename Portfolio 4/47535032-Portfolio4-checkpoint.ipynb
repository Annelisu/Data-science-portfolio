## 1. Linear Regression Modelling 
In this section, we will look into splitting and training the dataset, where the target variable will be the calories burnt. We will ensure that the features include both numerical and encoded categorical columns that are related to the target variable.


```python
# Convert categorical columns to numerical 
categorical_columns = ['Gender', 'Workout_Type']

# Initialise LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to categorical columns
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Features and target variable
X = data.drop(columns=['Calories_Burned', 'Gender', 'Workout_Type'])  
y = data['Calories_Burned']

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47535032)

print(X_train.shape)
print(X_test.shape)
```

```python
# Drop rows with null values for x and y
X_train = X_train.dropna()
X_test = X_test.dropna()
y_train = y_train[X_train.index]
y_test = y_test[X_test.index]

# fitting check
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Make predictions
y_pred = linear_model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R2:", r2)

```

### The Evaluation Metrics
* Mean Squared Error (MSE): 1762.55
* Root Mean Squared Error (RMSE): 41.98
* Mean Absolute Error (MAE): 32.70
* Coefficient of Determination (R²): 0.97

Given that the MSE, RMSE, and MAE are all very close to 0, and the coefficient of determination is close to 1, which means that the model fits the dataset well. While this test is not overfitting, I will add some test on new data to ensure that it generalises well.



## Polynomial Regression
To enhance the result, we will use the Polynomial Regression model. I will be applying 2 polynomial features for this case.

```python
# Set the degree 
degree = 2  
poly_model = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Fit and make prediction of the Regression model
poly_model.fit(X_train, y_train)
y_pred_poly = poly_model.predict(X_test)

# Calculate the evaluation metrics
mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = np.sqrt(mse_poly)
mae_poly = mean_absolute_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)


print("MSE (Polynomial):", mse_poly)
print("RMSE (Polynomial):", rmse_poly)
print("MAE (Polynomial):", mae_poly)
print("R2 (Polynomial):", r2_poly)
```

## Find the Important Feature using RFE Technique 
Next, we will use Recursive Feature Elimination (RFE) to determine the importance of features in predicting calories burned with a linear regression model. This helps identify which features are most influential in predicting members' calorie results.

```python
# Ensure only relevant features are included
X = data.drop(columns=['Calories_Burned']) 

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=47535032)

# Use RFE to rank features
rfe = RFE(estimator=LinearRegression(), n_features_to_select=1)
rfe.fit(X_train, y_train)

# Extract the features and place them in ranks
feature_ranking = pd.DataFrame({
    'Feature': X_train.columns,  
    'Rank': rfe.ranking_
}).sort_values(by='Rank')

feature_ranking
```

From this ranking table, 'Session duration' is the most influential feature in predicting calories burned, along with gender, height, and average BPM, indicating that the amount of time spent in a workout session, as well as other attributes like gender and height, play a crucial role in determining calories burned. Lower-ranking features like max BPM and resting BPM suggest that they are not strong predictors of calories burned.

## Clustering with K-means

```python
# Extract quantity and revenue for clustering
data_cluster = data[['Workout_Frequency (days/week)', 'Session_Duration (hours)']]

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_cluster)

# Use Elbow Method for optimal clustering
wcss = []  
cluster_range = range(1, 11)
for i in cluster_range:
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=47535032)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method 
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The elbow method aims to find the optimal number of clusters for K-means. From the graph, there is a sharp decrease in the within-cluster sum of squares (WCSS) starting from 2 to 3, indicating that the number of clusters should be either 2 or 3. I will use 2 clusters in this case to distinguish between members who train more intensively and those who engage in lighter workouts.

```python
# Applying k-means clustering with 2 clusters
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=47535032)
data['cluster'] = kmeans.fit_predict(scaled_data)

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Workout_Frequency (days/week)', y='Session_Duration (hours)', 
                hue='cluster', palette='magma', s=100, alpha=0.7)
plt.title('K-means Clustering: Workout Frequency vs Session Duration')
plt.xlabel('Workout Frequency (days/week)')
plt.ylabel('Session Duration (hours)')
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Analysis:
The clusters are well separated with no significant overlap, indicating a clear distinction between two types of workout routines. The plot displays two distinct clusters where the two groups exhibit different workout behaviours based on frequency. The purple cluster represents members with a higher workout frequency, approximately 40-50 sessions. This cluster is likely to represent highly experienced individuals who work out more frequently and for longer periods of time. Conversely, the red cluster represents members with a much lower workout frequency, estimated at about 20-30 sessions. This red cluster could indicate less experienced members who spend less time in the gym and work out less frequently.

### Conclusion:
Cluster 0 likely represents advanced members who maintain a higher frequency of workouts and are more committed to building their fitness routines, resulting in more frequent and longer workout sessions. On the other hand, cluster 1 represents beginner-level members who might have just started their workout journeys, working out less frequently compared to cluster 0. This clustering provides insightful data, which introduces to new marketing strategies to maintain members engagement and help them to reach new fitness goals.
